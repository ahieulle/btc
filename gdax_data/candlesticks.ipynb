```python
import time
import datetime
import pandas
from pymongo import MongoClient
from pandas import DataFrame, notnull, isnull
from gdax import WebsocketClient
```

```python
import matplotlib.pyplot as plt
from matplotlib.finance import candlestick_ohlc
import matplotlib.dates as mdates
```

```python
mongo_client = MongoClient('mongodb://localhost:27017/')
ltc_collection = mongo_client.crypto.ltc_eur
df = DataFrame([t for t in ltc_collection.find() if t["type"] == "ticker"])
df.index = pandas.to_datetime(df["time"])
s = df["price"][1:]
dohlc = s.resample("15T").agg(["first", "max", "min", "last"]).astype(float)
dohlc.columns = ["open", "high", "low", "close"]
dohlc['date'] = dohlc.index
#dohlc = dohlc.loc[:, ["date", "open", "high", "low", "close"]]
```

```python
dohlc = dohlc.loc[isnull(dohlc).sum(axis=1) == 0 , ["date", "open", "high", "low", "close"]]
```

```python
dohlc
```

```python
def bollinger_bands(prices, window, n_sigma=2):
    mu = prices.rolling(window=window).mean()
    sigma = prices.rolling(window=window).std()
    up = mu + n_sigma * sigma
    low = mu - n_sigma * sigma
    return mu, up, low
```

```python
BB13mu, BB13up, BB13low = bollinger_bands(dohlc["close"], 13)
```

```python
from math import pi

from bokeh.plotting import figure, show, output_file

inc = dohlc["close"] > dohlc["open"]
dec = dohlc["open"] > dohlc["close"]
w = 15*60*1000 # 15min in ms

TOOLS = "pan,wheel_zoom,box_zoom,reset,save"

p = figure(x_axis_type="datetime", tools=TOOLS, plot_width=1000, title = "Candlestick")
p.xaxis.major_label_orientation = pi/4
p.grid.grid_line_alpha=0.3

p.segment(dohlc["date"], dohlc["high"], dohlc["date"], dohlc["low"], color="black")
p.vbar(dohlc["date"][inc], w, dohlc["open"][inc], dohlc["close"][inc], fill_color="#00FF00", line_color="black")
p.vbar(dohlc["date"][dec], w, dohlc["open"][dec], dohlc["close"][dec], fill_color="#F2583E", line_color="black")
#p.line(dohlc["date"], dohlc["close"])

bb13mu, bb13up, bb13low = bollinger_bands(dohlc["close"], 13)

p.line(dohlc["date"], bb13mu)
p.line(dohlc["date"], bb13up)
p.line(dohlc["date"], bb13low)

output_file("candlestick.html", title="candlestick.py example")

show(p)  # open a browser
```
